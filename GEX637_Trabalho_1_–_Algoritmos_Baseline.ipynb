```python
import numpy as np
import pandas as pd
import time
import sys                                    
from itertools import combinations, chain, accumulate, permutations
from functools import lru_cache, reduce
import os

instances = {
    "dj38.tsp": "Djibouti",
    "qa194.tsp": "Qatar",
    "uy734.tsp": "Uruguay",
    "wi29.tsp": "Western Sahara",
    "zi929.tsp": "Zimbabwe",
}
path = '/content/drive/MyDrive/GEX637/'
def create_matriz_instance(filename):
  content = []
  with open(filename) as f:
    for line in f.read().splitlines():
      if line[0].isdigit():
        content.append(line.strip())
  length_tsp = content.__len__()
  dimension = np.ndarray((length_tsp, length_tsp), dtype=int)
  for key_x, x in enumerate(content):
    id_x, x1, y1 = np.array(x.split(), dtype=float)
    for key_y, y in enumerate(content):
      id_y, x2, y2 = np.array(y.split(), dtype=float)
      dimension[key_x][key_y] = round(np.sqrt((x1-x2)**2+(y1-y2)**2))
  return dimension, length_tsp

def run_bta(length_tsp, dimension):
  array_random = np.arange(length_tsp)
  elapsed_time_instance = length_tsp if length_tsp < 999 else 60
  elapsed_time = 0
  best_tsp = {'time': [], 'caminho': [], 't-medio': []}
  for _ in range(10):
    qualidade = float('inf')
    t = time.process_time()
    while(elapsed_time < elapsed_time_instance):
      distancia = 0
      np.random.shuffle(array_random)
      elapsed_time = time.process_time() - t
      for key, position in enumerate(array_random):
        try:
          distancia += dimension[position][array_random[key+1]]
        except:
          distancia += dimension[position][array_random[0]]
      if distancia < qualidade:
        qualidade = distancia
        best_tsp['time'].append(distancia)
        best_tsp['caminho'] = [array_random]
      best_tsp['t-medio'].append(time.process_time() - t)
    return best_tsp

def main():
  resultados = []
  for filename in os.scandir(path):
    dimension, length_tsp = create_matriz_instance(filename)
    best_tsp = run_bta(length_tsp, dimension)
    resultados.append({
        'instancia': instances[filename.name],
        'autoria': 'Diogo.Baltazar',
        'algoritmo': 'BTA',
        'q-medio': round(np.mean(best_tsp['time'])),
        'q-desvio': round(np.std(best_tsp['t-medio']), 2),
        't-medio': round(np.mean(best_tsp['t-medio']))
    })
    print(pd.DataFrame(resultados))
  pd.DataFrame(resultados).to_csv('resultados.csv', index=False)
main()
```

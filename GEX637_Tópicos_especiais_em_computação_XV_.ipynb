### Importando bibliotecas e abertura do arquivo

```python
import numpy as np
import math
import pandas as pd
import time
import sys


with open('wi29.tsp') as f:
  content = [line.strip() for line in f.read().splitlines()]
length_tsp = int(content[4].split()[-1].strip())+1
length_tsp 
```

### Removendo valores iniciais e EOF do final

```python
content = content[7:]
content = content[:-1]
content[-1]
```

## Criação da matriz com o calculo das distancias entre os pontos

```python
array_result = []
dimension = np.ndarray((length_tsp-1, length_tsp-1), dtype=int)
for key_x, x in enumerate(content):
  id_x, x1, y1 = np.array(x.split(), dtype=float)
  for key_y, y in enumerate(content):
    id_y, x2, y2 = np.array(y.split(), dtype=float)
    dimension[key_x][key_y] = round(math.sqrt(math.pow(x1-x2, 2)+math.pow(y1-y2, 2)))
df = pd.DataFrame(dimension, dtype=int)
df

```

### Verificando qual é a menor distancia a ser percorrida em X tempo

```python
t = time.process_time()
best_tsp = {'time': float('inf'), 'caminho': []}
array_random = np.arange(length_tsp-1)
elapsed_time = 0
while(elapsed_time < 1):
  distancia = 0
  np.random.shuffle(array_random)
  elapsed_time = time.process_time() - t
  for key, position in enumerate(array_random):
    try:
      distancia += dimension[position][array_random[key+1]]
    except:
      distancia += dimension[position][array_random[0]]
  if distancia < best_tsp['time']:
    best_tsp['time'] = distancia
    best_tsp['caminho'] = array_random

best_tsp
```

```python
import numpy as np
import math
import pandas as pd


with open('wi29.tsp') as f:
  content = [line.strip() for line in f.read().splitlines()]
length_tsp = int(content[4].split()[-1].strip())+1
length_tsp 
```

```python
content = content[7:]
content = content[:-1]
content[-1]
```

```python
array_result = []
dimension = np.array([np.zeros(length_tsp)])
for key_x, x in enumerate(content):
  id_x, x1, x2 = np.array(x.split(), dtype=float)
  for key_y, y in enumerate(content):
    id_y, y1, y2 = np.array(y.split(), dtype=float)
    result = math.sqrt(math.pow(x1-x2, 2)+math.pow(y1-y2, 2))
    dimension[key_x][key_y] = math.sqrt(math.pow(x1-x2, 2)+math.pow(y1-y2, 2))
  dimension = np.append(dimension, np.array([np.zeros(length_tsp)]), axis=0)
df = pd.DataFrame(dimension)


```
